```python
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

print('Dependencies installed!')
```

```python
# Reading and opening the file:
recent_grads = pd.read_csv('recent-grads.csv')

print('Displayinh the different categories:')
recent_grads.iloc[0]
```

```python
print('Displaying the first 5 rows of the dataset:')
recent_grads.head()
```

```python
print('Displaying the last 5 rows of the dataset:')
recent_grads.tail()
```

```python
print('Displaying the different averages collected from the dataset:')
recent_grads.describe()
```

```python
raw_data_count = recent_grads.columns
print('Printing all the rows in the dataset:')
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
print('Dropping all missing values from the dataset:')
recent_grads
```

## Pandas, Scatter plots:

```python
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_title('Sample Size VS. Median')
```

```python
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Sample Size VS. Unemployment Rate')
```

```python
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_title('Full Time VS. Median')
```

```python
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax4.set_title('Share Women VS. Unemployment Rate')
```

```python
ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax5.set_title('Men VS. Median')
```

```python
ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax6.set_title('Women VS. Median')
```
